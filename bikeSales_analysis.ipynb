```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the function to read all sheets from an Excel file
# This function will give us a dictionary of dataframes for each sheet

def read_all_sheets_from_excel(path: str) -> dict:
    return pd.read_excel(path, sheet_name=None)

# Read the Excel file
file_path = '/content/drive/MyDrive/Data analysis/Bikesale Analysis - bike_buyers.csv'
dataframes = pd.read_csv(file_path)
dataframes.head()

# Plot the distributions of the first few dataframes
for sheet, df in dataframes.items():
    print(f'--- {sheet} ---')
    print(df.head())
    df.hist(figsize=(10, 6))
    plt.suptitle(f'Distribution of {sheet}', y=1.02)
    plt.show()
```

Pairplot visualization of the numerical columns in the 'bike_buyers' sheet:


```python
import seaborn as sns
import pandas as pd

# Create a pairplot to visualize relationships between numerical columns
file_path = '/content/drive/MyDrive/Data analysis/Bikesale Analysis - bike_buyers.csv'
df = pd.read_csv(file_path)
sns.pairplot(df)
```

This plot provides a pairwise relationships in the dataset. The pairplot function creates a grid of Axes such that each variable in the data will be shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are histograms that show the distribution of a single variable. Off-diagonal plots are scatterplots that show the relationship (or lack thereof) between two variables.



# **Heatmap visualization of the correlation between numerical columns in the 'bike_buyers' sheet:**


```python
import seaborn as sns
import pandas as pd

# Load the DataFrame from the CSV file
file_path = '/content/drive/MyDrive/Data analysis/Bikesale Analysis - bike_buyers.csv'
df = pd.read_csv(file_path)

# Create a heatmap to visualize the correlation between numerical columns
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
```

This heatmap provides a color-coded representation of the correlation matrix of the data. The correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. A correlation of 1 means a perfect positive correlation, while a correlation of -1 means a perfect negative correlation. This can help us understand the relationship between different variables in the dataset.

# Plot showing the distribution of bike buyers by region:

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/Data analysis/Bikesale Analysis - bike_buyers.csv')

# Create a bar plot of the 'Region' column
region_counts = df['Region'].value_counts()
region_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Bike Buyers by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This plot provides a visual representation of the number of bike buyers in each region. It can help us understand which regions have the highest and lowest number of bike buyers.
